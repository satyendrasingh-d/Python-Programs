```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips)
```

```python
# Scatter Plot axis Lavel Function
sns.scatterplot(data=tips,x='total_bill',y='tip', color = 'red')
```

```python
# Rel Plot figure lavel function square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', color = 'yellow')
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```python
flights_wide = flights.pivot(index='year', columns='month', values='passengers')
flights_wide
```

```python
sns.lineplot(data=flights_wide['May'])
```

```python
sns.lineplot(data=flights_wide)
```

```python
# Seaborn Histplot
# A histogram is a classic visulaizition tool that represents the distribution of one or more variables by counting the numbers 
# of observations that fall within discrete bins.
```

```python
penguins = sns.load_dataset('penguins')
sns.histplot(data=penguins, x='flipper_length_mm')
```

```python
sns.histplot(data=penguins, y='flipper_length_mm', color='red')
```

```python
sns.histplot(data=penguins, x='flipper_length_mm', bins=30, color = 'yellow')
```

```python
#seaborn.kdeplot
#Plot univariate or bivariate distributions using kernel density estimation.
#A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.
#KDE represents the data using a continuous probability density curve in one or more dimensions.
```

```python
# Plot a univariate distribution along the x axis
tips = sns.load_dataset('tips')
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.kdeplot(data=tips, y='total_bill')
```

```python
iris = sns.load_dataset('iris')
sns.kdeplot(data=iris)
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='time', multiple='stack')
```

```python
#seaborn.boxplot
#A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between 
#variables or across levels of a categorical variable.df = sns.load_dataset("titanic")
titanic=sns.load_dataset("titanic")
titanic
```

```python
sns.boxplot(x=titanic['age'])
```

```python
sns.boxplot(data=titanic, x='class', y='age', hue='alive')
```

```python
# seaborn violinplot
# It show the distribution of data points after grouping by one (or more) variables.
sns.violinplot(x=titanic['age'])
```

```python
sns.violinplot(data=titanic, x='age', y='class', color= 'pink')
```

```python
sns.violinplot(data=titanic, x='class', y='age', hue='alive')
```

```python
#seaborn.barplot
#A bar plot represents an aggregate or statistical estimate for a numeric variable with the height of each rectangle
# and indicates the uncertainty around that estimate using an error bar.
penguins = sns.load_dataset("penguins")
sns.barplot(penguins, x="island", y="body_mass_g")
```

```python
sns.barplot(penguins, x='body_mass_g', y='island', hue='island', legend=False)
```

```python
# Seaborn countplot
# A count plot can be thousand of as a histogram across a categorical instesd of quantitative variable.
sns.countplot(titanic, x='class')
```

```python
sns.countplot(titanic, x='class', hue='survived')
```

```python
sns.countplot(titanic, x='class', hue='survived', stat='percent')
```
