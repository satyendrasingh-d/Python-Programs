```python
# Matplotlib Scatter
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([66,85,87,88,111,86,183,87,94,78,77,85,86])
plt.scatter(x, y)
plt.show()
```

```python
# Matplotlib Bars
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['A','B','C','D','E'])
y = np.array([8,5,3,7,9])
plt.bar(x,y)
plt.show()
```

```python
# Matplotlib Bars "Color"
import matplotlib.pyplot as plt
import numpy as np
a = np.array(['A','B','C','D','E','F'])
b = np.array([18,15,35,25,14,50])
plt.bar(a,b, color = "Orange")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['A','B','C','D'])
y = np.array([30,10,25,45])
plt.bar(x,y, width = 0.10, color = "Green")
plt.show()
```

```python
# Matplotlib Histogram
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(250,20,200)
plt.hist(x, color = "Red")
plt.show()
```

```python
# Matplotlib Pie Charts
a = np.array([25,40,15,10,5])
plt.pie(a)
plt.show()
```
